```python
## Day 4 - Grouping and Aggregation
```

```python
# Average bill by day
```

```python
df.groupby("day")["total_bill_gbp"].mean()
```

```python
# Average tip by time 
```

```python
df.groupby("time")["tip"].mean()
```

```python
## Sunday has the highest average total bill. Dinner is more profitable tipping wise
```

```python
df.groupby("day").agg(
    avg_bill=("total_bill_gbp", "mean"),
    avg_tip=("tip", "mean"),
    count=("total_bill_gbp", "count")
)
```

```python
## Count is important as it allows us to study which days are busiest. Saturday looks the most reliable as it yields the greatest earinings
```

```python
df.groupby(["day", "time"])["total_bill_gbp"].mean()
```

```python
df.groupby(["day", "smoker"])["tip"].mean()
```

```python
## Significantly lower mean bills during lunchtime friday compares to every other time. I would ask smoker average bills grouped by day and time
```

```python
## Day 6 - Derived Columns & Feature Engineering
```

```python
df["tip_pct"] = df["tip"] / df["total_bill_gbp"] * 100
```

```python
df.head()
```

```python
df["tip_pct"].describe()
```

```python
### Average tip percentage = 16%. Max tip is so much higher than the lowest tip (71% v 3.5%)
```

```python
df.groupby("day")["tip_pct"].mean()
```

```python
df.groupby("smoker")["tip_pct"].mean()
```

```python
### Smokers tip slightly more percentage wise than non-smokers. Compared with raw tip values, they match up quite similar 
```

```python
df["bill_size"] = pd.cut(
    df["total_bill_gbp"],
    bins=[0, 10, 20, 40, 100],
    labels=["Small", "Medium", "Large", "Very Large"]
)
```

```python
df["bill_size"].value_counts()
```

```python
### Medium size bills, as I wouldve expected, are the most common. Categorising like this helps make more agruments and conclusions with regards to data
```

```python
df.groupby("bill_size")["tip_pct"].mean()
```

```python
### Larger bills tend to tip a lower percentage. This may be because the total bill price being high would inadvertively make the tip seem smaller in contrast, as opposed to paying the same tip with a smaller bill
```

```python
## Day 7 - Data Visualisation Basics
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["total_bill_gbp"], bins=20)
plt.title("Distribution of Total Bills")
plt.xlabel("Total Bill (GBP)")
```

```python
### The distribution is skewed to the left. Most bills cluster between 10 and 30 GBP
```

```python
sns.barplot(x="day", y="tip_pct", data=df)
plt.title("Average Tip Percentage by Day")
plt.ylabel("Tip Percentage")
```

```python
### Friday has the highest average tip
```

```python
sns.barplot(x="bill_size", y="tip_pct", data=df)
plt.title("Tip Percentage by Bill Size")
plt.ylabel("Tip Percentage")
```

```python
### Larger bills do not proportionally tip more. This could be because if both people with small and large bills tip the same, it would look smaller proportionally with larger bills as opposed to smaller bills
```
