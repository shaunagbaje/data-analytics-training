```python
## Day 4 - Grouping and Aggregation
```

```python
# Average bill by day
```

```python
df.groupby("day")["total_bill_gbp"].mean()
```

```python
# Average tip by time 
```

```python
df.groupby("time")["tip"].mean()
```

```python
## Sunday has the highest average total bill. Dinner is more profitable tipping wise
```

```python
df.groupby("day").agg(
    avg_bill=("total_bill_gbp", "mean"),
    avg_tip=("tip", "mean"),
    count=("total_bill_gbp", "count")
)
```

```python
## Count is important as it allows us to study which days are busiest. Saturday looks the most reliable as it yields the greatest earinings
```

```python
df.groupby(["day", "time"])["total_bill_gbp"].mean()
```

```python
df.groupby(["day", "smoker"])["tip"].mean()
```

```python
## Significantly lower mean bills during lunchtime friday compares to every other time. I would ask smoker average bills grouped by day and time
```
